```python
import numpy as np             # import some python libraries
import pandas as pd 
```

```python
train = pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")  # importing training dataset
train
```

```python
train.drop(['Name'],axis=1,inplace=True) #dropping name
```

```python
train['Sex']=train['Sex'].map({"female":0,"male":1}) # Assigning values
```

```python
train['Age'].fillna(value=train['Age'].mean(), inplace=True) # Replacing null values with mean
```

```python
train['Fare']=train['Fare'].round(2)  # Rounding Fare upto 2nd decimal value
```

```python
train['Age']=train['Age'].round(2)   # Rounding Age upto 2nd decimal value
```

```python
Cab=train['Cabin'].unique()    # Checking out unique values of Cabin Column in Training Dataset
print(Cab)
```

```python
print(len(train['Cabin'].unique()))  # Total unique values in Cabin Column in training dataset
```

```python
f=dict()
count=1
for a in train['Cabin'].unique():              ## Assigning numerical values to Cabin Column
    f[a]=count/120
    print("a:", a , "count:",count,",f[a]:",f[a])
    count+=1
train['Cabin']=train['Cabin'].map(f)
```

```python
train['Cabin']=train.Cabin.round(4)    # Rounding all Cabin Column values to 4th decimal value.
```

```python
train['Embarked'].unique()     # Checking all unique values in Column Embarked for Training Dataset.
```

```python
d=dict()
count=1
for i in train['Embarked'].unique():        # Assigning numerical values to Embarked Column. 
    d[i]=count/4
    print("i:", i , "count:",count,",d[i]:",d[i])
    count+=1
train['Embarked']=train['Embarked'].map(d)
```

```python
train.drop(['Ticket'],axis=1,inplace=True) #dropping ticket column
```
